In Changeup, Check Pearson R2: if other parameters were predictable from spin rate to speed:
    
I found speed actually has the highest correlation from other parameters (Vy0 is high, but Vy0 means the velocity of y-axis, means if the speed has a high R2, the Vx0 or Vy0 or Vz0 will also be high)

````python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from yellowbrick.target import FeatureCorrelation


Richard_CH=Richard[Richard.pitch_name=='Changeup']
target=Richard_CH['release_spin_rate']
features=Richard_CH.drop('release_spin_rate',axis=1)
features=features.drop(['game_date', 'pitch_name'], axis=1)

visualizer = FeatureCorrelation(labels=list(features.columns))
visualizer.fit(features, target)
visualizer.poof()
````

So I start to do linear regression analysis, based on the drop time: `2021/6/11` to separate the dataset

````python
Richard_CH_sample_yes = Richard_CH[Richard_CH.game_date >= '2021/6/11' ]
Richard_CH_sample_no = Richard_CH[Richard_CH.game_date <= '2021/6/11' ]
````

In order to balance the two different separated datasets, I need to know how much count each datasets has

````python
#collapse-output
Richard_CH_sample_yes.info()
````

````python
#collapse-output
Richard_CH_sample_no.info()
````

Sample the data and define the order and then plot order with actual value to check if it is a normal distribution

````python
Richard_CH_sample_yes = Richard_CH_sample_yes.sample(19)
from sklearn.preprocessing import scale
diff_res = scale(np.array(Richard_CH_sample_yes['release_spin_rate']) - np.array(Richard_CH_sample_no['release_spin_rate']))
plt.hist(diff_res)
````

Even the dataset is only `19` counts, it still can tell me it is not a normal distribution. Let's check the QQ plot.
the result tells me there is an bimodal distribution, which is not a normal distribution

````python
from scipy import stats
stats.probplot(diff_res, plot=plt, dist='norm');
````

Calculate the Levene and found the significance P-value<0.05, proved after the crackdown, the quality of pitching is different from the before. 

````python
stats.levene(Richard_CH_sample_yes['release_spin_rate'], Richard_CH_sample_no['release_spin_rate'])
````
